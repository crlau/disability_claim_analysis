```python
# Using Python 3.6.4
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

# returning Pandas dataframe object from CSV dataset
disability_df = pd.read_csv('SSA-SA-FYWL.csv')
```

This bar graph shows the distribution of SSA Disability Beneficiaries (age 18-64) by state in 2015. It shows that CA, NY, TX, and FL have the most number of SSA Disability Beneficiaries.

```python
# isolating 2015 data from all states
data_2015 = disability_df.loc[lambda df: disability_df['Date'] == 2015, :]

# converting data series to list type
states = data_2015['State Code'].tolist()
state_ben = data_2015['SSA Disabilty Beneficiaries  age 18-64'].tolist()
state_ben = list(map(int, state_ben))

# creating bar graph
x = np.arange(52)
plt.bar(x, state_ben, align='center')
plt.xticks(x, states)
plt.title('Distribution of SSA Disability Beneficiaries (age 18-64) in the US in 2015')
plt.ylabel('Number of SSA Disability Beneficiaries')
plt.xlabel('State')

# changing the size of the canvas to fit all bars
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 80
fig_size[1] = 40
plt.rcParams["figure.figsize"] = fig_size
plt.tight_layout()

# showing the plot
plt.show()
```

This boxplot shows the number of states that have a disproportionate amount of SSA Disability Beneficiaries in the year 2015.

```python
# States that have outlier numbers of SSA Disability Beneficiaries in 2015
plt.boxplot(state_ben)
plt.title('Outlier Visualization of SSA Disability Beneficiaries in 2015')
plt.xlabel('Distribution')
plt.ylabel('Number of SSA Disability Beneficiaries')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 5
plt.rcParams["figure.figsize"] = fig_size
plt.show()
```

This scatterplot shows the increase of the number of SSA Beneficiaries (age 18-64) from 2001-2015 in North Carolina. 

```python
# Isolation of data to North Carolina (2001-2015)
nc_data = disability_df.loc[lambda df: disability_df['State Code'] == 'NC ', :]

# converting data series desired to list type
years = nc_data['Date'].tolist()
nc_ben_data = nc_data['SSA Disabilty Beneficiaries  age 18-64'].tolist()
nc_ben_data = list(map(int, nc_ben_data))

# creating the scatter plot
plt.scatter(years, nc_ben_data, color='orange')
plt.title('Trend of SSA Disability Beneficiaries(age 18-64) 2001-2015 in North Carolina')
plt.ylabel('Number of SSA Disability Beneficiaries')
plt.xticks(range(2001, 2016, 1))

plt.show()
```

These two subplots compare the trend of SSA Disability Beneficiaries (2001-2015) for the states of California and North Carolina. This shows that both states share a similar trend of increase from 2001-2013 and a tapering off from 2013-2015. California's number of beneficiaries is larger, but may be due to a larger population that inhabits the state.

```python
# Isolation of data to California
ca_data = disability_df.loc[lambda df: disability_df['State Code'] == 'CA ']
ca_ben_data = ca_data['SSA Disabilty Beneficiaries  age 18-64'].tolist()
ca_ben_data = list(map(int, ca_ben_data))

# California plot
plt.subplot(2,1,1)
plt.plot(years, ca_ben_data, color='navy')
plt.title('Trend of SSA Disability Beneficiaries 2001-2015 in California')
plt.xlabel('Year')
plt.ylabel('Number of SSA Disability Beneficiaries')
plt.xticks(range(2001, 2016, 1))

# North Carolina plot
plt.subplot(2,1,2)
plt.plot(years, nc_ben_data, color='orange')
plt.title('Trend of SSA Disability Beneficiaries 2001-2015 in North Carolina')
plt.xlabel('Year')
plt.ylabel('Number of SSA Disability Beneficiaries')
plt.xticks(range(2001, 2016, 1))

# changing size of plot
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 10
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
plt.tight_layout()

plt.show()
```
